# Setting Up 

To conduct this analysis, we’ll be using FsLab, a conglomerate of Data Science libraries that makes data access, analysis and visualization using F# extremely simple.
The result of the Data Acquisition process was a CSV file with all the movie data that I load into a Deedle Data Frame that’s akin to a data frame in the R and Python world. Deedle is an exceptional Exploratory Data library developed by the good people of Blue Mountain Capital and other contributors.



```
// Dependencies
#r "nuget: Deedle"
#r "nuget: XPlot.Plotly"

// Imports 
open XPlot.Plotly
open Deedle
open System
```

```
[<Literal>]
let moviesFile =  __SOURCE_DIRECTORY__ +  @"/data/Movies.csv"

let moviesDf : Frame< int, string > = Frame.ReadCsv( moviesFile )

// Setting the Index of the movies data frame to that of the name of the movie / series
let indexedMoviesDf = 
    moviesDf
    |> Frame.indexRowsString "Name"

let movieSeriesDf =
    indexedMoviesDf
    |> Frame.sliceRows[ "The Lord of the Rings Series"; "The Hobbit Series" ]

moviesDf.Print()
```

## Feature 1: Average Score from Rotten Tomatoes

```
// Feature 1: Rotten Tomatoes 
let rottenTomatoesScoreDf = 
    movieSeriesDf
    |> Frame.sliceCols [ "RottenTomatoesScore";  ]

// Chart the Table of the Rotten Tomatoes Scores Data Frame
display(rottenTomatoesScoreDf)
```
